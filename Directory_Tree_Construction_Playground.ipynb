# GCA Dataset Summary

```python
import os
import shutil
import glob
import time
```

```python
DATA_DIR = fr"F:\GCA\ND-Annotations\ALL_Labelled"
print(DATA_DIR)
```

```python
dir_list = os.listdir(DATA_DIR)
categories = ["N", "E", "NAR", "Y"]
# print(dir_list)
```

```python
for i, j in enumerate(os.listdir(DATA_DIR)[0:2]):
  print(f"{i} {j}")
```

```python
start = time.time()
# delete 00-Model_DATA if you're re-creating directory
if os.path.exists(f"{DATA_DIR}\\00-Model_DATA\\") and os.path.isdir(f"{DATA_DIR}\\00-Model_DATA\\"):
  shutil.rmtree(f"{DATA_DIR}\\00-Model_DATA\\")
for folder_count, img_folder in enumerate(os.listdir(DATA_DIR)):
  print(f"\n\ncount:{folder_count+1}\nname:{img_folder}\n")
  for category in categories:
    for file in glob.glob(f"{DATA_DIR}\{img_folder}\{category}\*"):
      destination = f"{DATA_DIR}\\00-Model_DATA\\{category}\\"
      os.makedirs(os.path.dirname(destination), exist_ok=True)
      new_path = shutil.copy(file, destination)
      print(new_path)
print(f"\nTotal number of folders: {folder_count+1}")
end = time.time()
delta = end - start
print(f"Time taken: {round(delta, 2)} seconds")
```
